```python
%matplotlib inline

#Librerías
import numpy as np
import matplotlib.pyplot as plt

#Plano Cartesiano

ancho = 10 
alto = 10 

# Figura y ejes
fig, ax = plt.subplots(figsize=(ancho, alto))

# Longitud de ejes
xmin, xmax, ymin, ymax = -5, 5, -5, 5

# Escalado identico para los ejes
ax.set(
    # eje horizontal
    xlim=(xmin, xmax), 
    # eje vertical
    ylim=(ymin, ymax), 
    # mantener aspecto igual en ambos ejes
    aspect='equal'
)
# Posiciona las rectas izquierda y derecha como X y Y en un sistema de coordenadas en este caso cero
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

# Cambia de color las rectas
ax.spines['bottom'].set_color('r')
ax.spines['left'].set_color('b')


# Elimina las líneas superior y derecha de la figura
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Crear las etiquetas para 'x' and 'y' localizadas al final de los ejes
ax.set_xlabel('α', size=20, labelpad=-24, x=1.03)
ax.set_ylabel('β', size=20, labelpad=-21, y=1.02, rotation=0)

# Espacio entre las etiquetas de marca
ticks_frequency = 0.5

# Crear marcas (mayores) para determinar la posición de las etiquetas de marca
x_ticks = np.arange(xmin, xmax+.5, ticks_frequency)
y_ticks = np.arange(ymin, ymax+.5, ticks_frequency)
ax.set_xticks(x_ticks[x_ticks != 0])
ax.set_yticks(y_ticks[y_ticks != 0])

# Crear marcas menores localizadas en cada entero para permitir dibujar una cuadricula (menor)
# Nota: En este ejemplo usando ticks_frequency=1, se nota el efecto
ax.set_xticks(np.arange(xmin, xmax+1), minor=True)
ax.set_yticks(np.arange(ymin, ymax+1), minor=True)

# Dibujar cuadricula
ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

# Dibujar flechas en los ejes
arrow_fmt = dict(markersize=4, color='orange', clip_on=False)
ax.plot((1), (0), marker='>', transform=ax.get_yaxis_transform(), **arrow_fmt)
arrow_fmt = dict(markersize=4, color='cyan', clip_on=False)
ax.plot((0), (1), marker='^', transform=ax.get_xaxis_transform(), **arrow_fmt)


plt.show()


```

```python
%matplotlib inline

#Librerías
import numpy as np
import matplotlib.pyplot as plt

ancho = 10 
alto = 10
xmin = -5
xmax = 5
ymin = -5
ymax = 5


def Plano(xmin,xmax,ymin,ymax,ancho,alto):

  # Figura y ejes
  fig, ax = plt.subplots(figsize=(ancho, alto))

  # Escalado identico para los ejes
  ax.set(
      # eje horizontal
      xlim=(xmin, xmax), 
      # eje vertical
      ylim=(ymin, ymax), 
      # mantener aspecto igual en ambos ejes
      aspect='equal'
  )
  # Posiciona las rectas izquierda y derecha como X y Y en un sistema de coordenadas en este caso cero
  ax.spines['bottom'].set_position('zero')
  ax.spines['left'].set_position('zero')

  # Cambia de color las rectas
  ax.spines['bottom'].set_color('r')
  ax.spines['left'].set_color('b')


  # Elimina las líneas superior y derecha de la figura
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)

  # Crear las etiquetas para 'x' and 'y' localizadas al final de los ejes
  ax.set_xlabel('α', size=20, labelpad=-24, x=1.03)
  ax.set_ylabel('β', size=20, labelpad=-21, y=1.02, rotation=0)

  # Espacio entre las etiquetas de marca
  ticks_frequency = 0.5

  # Crear marcas (mayores) para determinar la posición de las etiquetas de marca
  x_ticks = np.arange(xmin, xmax+.5, ticks_frequency)
  y_ticks = np.arange(ymin, ymax+.5, ticks_frequency)
  ax.set_xticks(x_ticks[x_ticks != 0])
  ax.set_yticks(y_ticks[y_ticks != 0])

  # Crear marcas menores localizadas en cada entero para permitir dibujar una cuadricula (menor)
  # Nota: En este ejemplo usando ticks_frequency=1, se nota el efecto
  ax.set_xticks(np.arange(xmin, xmax+1), minor=True)
  ax.set_yticks(np.arange(ymin, ymax+1), minor=True)

  # Dibujar cuadricula
  ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

  # Dibujar flechas en los ejes
  arrow_fmt = dict(markersize=4, color='orange', clip_on=False)
  ax.plot((1), (0), marker='>', transform=ax.get_yaxis_transform(), **arrow_fmt)
  arrow_fmt = dict(markersize=4, color='cyan', clip_on=False)
  ax.plot((0), (1), marker='^', transform=ax.get_xaxis_transform(), **arrow_fmt)

  plt.show()
  

Plano(xmin,xmax,ymin,ymax,ancho,alto)

```

```python
# Librerías
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

#Plano Cartesiano
ancho = 10
alto = 10

fig, ax = plt.subplots(figsize=(ancho, alto))
xmin, xmax, ymin, ymax = -10, 10, -10, 10
ax.set(xlim=(xmin, xmax), ylim=(ymin, ymax), aspect='equal')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_xlabel('x', size=14, labelpad=-24, x=1.03)
ax.set_ylabel('y', size=14, labelpad=-21, y=1.02, rotation=0)

ticks_frequency = 1
x_ticks = np.arange(xmin, xmax+1, ticks_frequency)
y_ticks = np.arange(ymin, ymax+1, ticks_frequency)
ax.set_xticks(x_ticks[x_ticks != 0])
ax.set_yticks(y_ticks[y_ticks != 0])

ax.set_xticks(np.arange(xmin, xmax+ 1), minor=True)
ax.set_yticks(np.arange(ymin, ymax+ 1), minor=True)

ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

arrow_fmt = dict(markersize=4, color='black', clip_on=False)
ax.plot((0), (1), marker='^', transform=ax.get_xaxis_transform(), **arrow_fmt)
ax.plot((0), (0), marker='v', transform=ax.get_xaxis_transform(), **arrow_fmt)
ax.plot((1), (0), marker='>', transform=ax.get_yaxis_transform(), **arrow_fmt)
ax.plot((0), (0), marker='<', transform=ax.get_yaxis_transform(), **arrow_fmt)

#Definicion de Puntos
punto1 = [9, 9]
punto2 = [-9, -9]

punto3 = [8, -8]
punto4 = [-8, 8]

#Definicion Abscisas y ordenadas
x_values = [punto1[0], punto2[0]]
y_values = [punto1[1], punto2[1]]

x2_values = [punto3[0], punto4[0]]
y2_values = [punto3[1], punto4[1]]


#Impresion de Puntos en el Plano
plt.plot(x_values, y_values,'bo',c='red',mfc='blue', mec= 'k', linestyle="--")
plt.plot(x2_values, y2_values,'bD',c='cyan',mfc='brown', mec= 'k', linestyle="--")

#Agregar etiquetas a los puntos
plt.text(punto1[0]-0.015, punto1[1]-0.45, "A(9,9)")
plt.text(punto2[0]+0.15, punto2[1]-0.35, "B(-9,-9)")

plt.text(punto3[0]-0.015, punto3[1]-0.45, "C(8,-8)")
plt.text(punto4[0]+0.15, punto4[1]+0.25, "D(-8,8)")

# Dibujar líneas de conexión entre los puntos y los ejes
colors = ['m', 'g', 'r', 'b']
for x, y, c in zip(x_values, y_values, colors):
    ax.plot([x, x], [0, y], c=c, ls='--', lw=1.5, alpha=0.5)
    ax.plot([0, x], [y, y], c=c, ls='--', lw=1.5, alpha=0.5)

colors = ['y', 'b', 'y', 'y']
for x, y, c in zip(x2_values, y2_values, colors):
    ax.plot([x, x], [0, y], c=c, ls='--', lw=1.5, alpha=0.5)
    ax.plot([0, x], [y, y], c=c, ls='--', lw=1.5, alpha=0.5)


plt.show()
```
